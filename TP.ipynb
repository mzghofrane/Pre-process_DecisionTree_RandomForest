# 2. Nettoyage des données

```python
#renseigner les valeurs manquantes
cat_data=[]
num_data=[]
for i,c in enumerate (df.dtypes):
     if c == 'object':
        cat_data.append(df.iloc[:,i])  # Ajoutez le nom de la colonne aux données catégorielles
     else:
        num_data.append(df.iloc[:,i])  # Ajoutez le nom de la colonne aux données numériques
```

```python
cat_data=pd.DataFrame(cat_data).transpose()
cat_data
```

```python
num_data=pd.DataFrame(num_data).transpose()
num_data
```

Q4) Quel est l'objectif de ces trois codes? est-t-il necessaire de faire cette étape avant de gérer les données manquantes?

Réponce : 

L'objectif de ces trois codes :
** Pour gérer les valeurs manquantes dans les variables catégoriques
** Remplacer les valeurs manquantes par la valeur la plus fréquente (mode) est une approche raisonnable. 

Il necessaire de faire cette étape avant de gérer les données manquantes par ce que cela peut aider à préserver la distribution des données et à éviter d'introduire des biais.

Pour les variables categoriques manquantes on va  les remplacer par les valeurs qui se répetent le plus

Q5) est ce qu'il ya une autre alternative 

Réponce : 
Alors pour gérer les valeurs manquantes dans les variables catégoriques, il existe plusieurs alternatives : 
*** On peut remplacer les valeurs manquantes par la valeur la plus fréquente
*** Créer une nouvelle catégorie pour les valeurs manquantes, 

```python
cat_data = cat_data.apply(lambda x: x.fillna(x.value_counts().index[0]))
cat_data.isnull().sum().any()
```

Pour les variables numériques manquantes on va remplacer les variables manquantes par la valeur précedente de la meme colonne

```python
num_data.fillna(method='bfill', inplace=True)
num_data.isnull().sum().any()
```

```python
num_data.bfill(inplace=True)
num_data.isnull().sum().any()
```

Transformation de la colonne cible "Loan_Status" en une variable numérique et la nommer " target" puis supprimer la colone Loan_status
Pour supprimer une colonne, on utilise la commande "drop". Le paramètre axis=1 indique qu'on veut supprimer une colonne, et 'inplace=True' signifie que la modification doit être effectuée directement sur le DataFrame lui même.

```python
target=cat_data['Loan_Status']
cat_data.drop('Loan_Status',axis=1,inplace=True)
target=target.map({'Y': 1, 'N': 0})
target
```

Remplacer les valeurs categoriques par des valeurs numeriques

```python
le=LabelEncoder()
for i in cat_data:
    cat_data[i]=le.fit_transform(cat_data[i])
cat_data
```

Regrouper les données numérique et catégoriques

```python
df=pd.concat([cat_data,num_data,target],axis=1)
```

```python
df
```

# Préparation des données pour l'apprentissage

Division des données en train et test 

random_state : une valeur entière qui détermine la reproductibilité des résultats. Fixer cette valeur garantit que chaque fois que vous exécutez le code, vous obtenez la même division des données.

stratify : spécifie la colonne à utiliser pour la stratification lors de la division des données. La stratification garantit que la répartition des classes dans l'ensemble d'entraînement et dans l'ensemble de test est similaire à celle du DataFrame d'origine, en particulier important pour les ensembles de données déséquilibrés.

```python
dfTrain, dfTest = train_test_split(df, test_size=300, random_state=1, stratify=df.Loan_Status)
print(dfTrain.shape)
print(dfTest.shape)
```

```python
print ("distribution relative des classes en apprentissage")
print (dfTrain.Loan_Status.value_counts(normalize=True))
print ("distribution relative des classes en test")
print (dfTest.Loan_Status.value_counts(normalize=True))
```
